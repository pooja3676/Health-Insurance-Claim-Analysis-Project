```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
column_name = df.columns

print(column_name)
```

```python
print("The Average value of Claim :" ,np.mean(df['Claim']))
print("The Minimum value of Health Insurance Claim :" ,np.min(df['Claim']))
print("The Maximum value of Health Insurance Claim :" ,np.max(df['Claim']))
```

```python
df['Claim'].value_counts().nlargest(11).plot(kind='bar')
plt.title('Top 10 Number of Total Claim')
plt.xlabel('Claim')
plt.ylabel('Count')
plt.show()
```

```python
In this Bar Chart, The Highest Health Insurance Claim is 14210.5 
and The Lowest Health Insurance claim is 8283.7.
```
